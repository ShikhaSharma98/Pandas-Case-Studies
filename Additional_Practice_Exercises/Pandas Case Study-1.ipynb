```python
#4. Create two subsets with unique and duplicate values. 

unique_values= cust_data[cust_data.Customer_ID.duplicated()==False]
duplicate_values= cust_data[cust_data.Customer_ID.duplicated()==True]
```

```python
#5. Create data set with list of customers whose customer value greater than 10000. 
```

```python
cust=cust_data[cust_data.Customer_Value>10000]
cust.head()
```

```python
#6. In customer table, create a new variable called “customer value segment” using customer value as follows. -  
#- High Value Segment - > 25000  
#- Medium Value Segment – Between 10000 and 25000  
#- Low Value Segment – less than or equal to 10000 
```

```python

cust_data['Customer value segement']=pd.cut(cust_data.Customer_Value,[0,10000,25000,max(cust_data.Customer_Value)],labels=['Low Value Segment','Medium Value Segment',' High Value Segment'])
cust_data.tail()
```

```python
#7. Create variables “average revenue per trip” and “balance points” in the customer data set.  
```

```python
cust_data=cust_data.assign(average_revenue_per_trip=cust_data.Customer_Value/cust_data.Buy_Times,balance_points=cust_data.Points_Earned-cust_data.Points_Redeemed)
cust_data.head(10)
```

```python
#8. How many days between last purchase date and today?
```

```python
last_purchase_date=cust_data.Recent_Date.max()
today=pd.datetime.now()
(today-last_purchase_date).days
```

```python
#9. Calculate percentage of sales by each last city, state and region.
```

```python
data=cust_data[['Last_region','Last_state','Last_city','Customer_Value']].groupby(['Last_region','Last_state','Last_city']).sum()
data.groupby(level=0).apply(lambda x: x/x.sum() * 100)
```

```python
#10. What is the count of customers, average number of purchases and average purchase transaction value by last state 
#and city. 
```

```python
cust_data[['Last_state','Last_city','Buy_Times','Customer_Value','Customer_ID',]].groupby(['Last_state','Last_city']).agg({'Customer_ID':'count','Customer_Value':'mean','Buy_Times':'mean'})
```
