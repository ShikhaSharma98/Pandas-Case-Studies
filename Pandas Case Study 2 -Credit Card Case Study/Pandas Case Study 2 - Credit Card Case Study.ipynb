```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cust_acq=pd.read_csv("Customer Acqusition.csv")
cust_acq.head()
```

```python
repayment=pd.read_csv("Repayment.csv")
repayment.head()
repayment.drop(['Unnamed: 4'],axis=1,inplace=True)
repayment.head()
```

```python
spend=pd.read_csv("spend.csv")
spend.head()
```

### 1. In the above dataset,
#### a. In case age is less than 18, replace it with mean of age values.

```python
cust_acq.loc[(cust_acq.Age<18),['Age']]=cust_acq.Age.mean()
```

#### b. In case spend amount is more than the limit, replace it with 50% of that customerâ€™s limit.

```python
cust_spend=pd.merge(cust_acq,spend,on='Customer',how='inner')
cust_spend['Amount']=np.where(cust_spend.Amount>cust_spend.Limit,cust_spend.Limit/2,cust_spend.Amount)
cust_spend.head(30)
```

#### c. Incase the repayment amount is more than the limit, replace the repayment with the limit.

```python
cust_repay=pd.merge(cust_acq,repayment,how='inner',on='Customer')
cust_repay['Amount']=np.where(cust_repay.Amount>cust_spend.Limit,cust_spend.Limit,cust_repay.Amount)
cust_repay.head(30)
```

#### c. What is the average monthly spend by customers?


```python
cust_spend.Month=pd.to_datetime(cust_spend.Month)
cust_spend['spend_month']=cust_spend['Month'].apply(lambda x:x.month)
monthly_spend_avg=cust_spend.groupby(['Customer','spend_month'])['Amount'].mean().reset_index()
monthly_spend_avg.head(20)
```

#### d. What is the average monthly repayment by customers?

```python
cust_repay.Month=pd.to_datetime(cust_repay.Month)
cust_repay['repay_month']=cust_repay['Month'].apply(lambda x:x.month)
monthly_repayment_avg=cust_repay.groupby(['Customer','repay_month'])['Amount'].mean().reset_index()
monthly_repayment_avg.head(20)
```

#### e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? 

```python
MonthlyCustomers_avg=pd.merge(monthly_spend_avg,monthly_repayment_avg,left_on=['Customer','spend_month'],right_on=['Customer','repay_month'],how='inner',suffixes=['_spend','_repayment'])
MonthlyCustomers_avg['Monthly_profit']=MonthlyCustomers_avg.Amount_repayment-MonthlyCustomers_avg.Amount_spend
MonthlyCustomers_avg['profit']=np.where(MonthlyCustomers_avg.Monthly_profit>0,MonthlyCustomers_avg.Monthly_profit*(2.9/100),np.nan)
MonthlyCustomers_avg.head(20)                                    
```

### 5. Write user defined PYTHON function to perform the following analysis:

#### You need to find top 10 customers for each city in terms of their repayment amount by different products and by different time periods i.e. year or month. The user should be able to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the function should automatically take these inputs while identifying the top 10 customers.

```python
cust_repay['yearly']=cust_repay.Month.apply(lambda x:x.year)
cust_repay['monthly']=cust_repay.Month.apply(lambda x:x.month)

def top10Customers(product_category,time_period):
    return cust_repay.loc[(cust_repay.Product == product_category)].groupby(['Customer','City','Product',time_period]).Amount.sum().reset_index().sort_values('Amount',ascending=False).head(10)
```

```python
product_category=str(input("Please Enter Product Category and product category should be in Gold/Silver/Platinum: "))
time_period=str(input("Please Enter Time Period and time period should be in yearly/monthly: "))
```

```python
top10Customers_Analysis=top10Customers(product_category,time_period)
top10Customers_Analysis
```
